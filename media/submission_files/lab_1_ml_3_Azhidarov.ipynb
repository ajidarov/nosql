### 6. Image Thresholding
1. To perform simple thresholding, use cv2.threshold() function, where first argument is the source image, which should be a grayscale image. Second argument is the threshold value which is used to classify the pixel values. Third argument is the maxVal which represents the value to be given if pixel value is more than (sometimes less than) the threshold value. And fourth argument is style of thresholding, use all of given below.
    
    Note:
    - cv2.THRESH_BINARY
    - cv2.THRESH_BINARY_INV
    - cv2.THRESH_TRUNC
    - cv2.THRESH_TOZERO
    - cv2.THRESH_TOZERO_INV
2. Make a list of names for resulting images.
3. Make a list of images.
4. Iterate over the list for displaying images.
5. Explain Image Threasholding process.

```python
# cv2.imshow('graaayy', res)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
gray = cv2.imread('yoda.jpg',0)
ret,thresh1 = cv2.threshold(gray,64,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(gray,64,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(gray,64,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(gray,64,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(gray,64,255,cv2.THRESH_TOZERO_INV)
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [gray, thresh1, thresh2, thresh3, thresh4, thresh5]

plt.subplots(2,3,figsize=(15,15))
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray',vmin=0,vmax=255)
    plt.title(titles[i])
plt.show()
```

<h3 style="color: gray">In my understanding, thresholding is used to filter out unneccecary data from the image, or to highlight some features of the image</h3>
<p style="color: gray">Threshold sets the lower and upper boundaries for the each pixel (64 and 255 for example, anything dimmer than 64 is not included, and 255 is max val)</p>

### 7. Image Smooting
1. Perform averaging. It simply takes the average of all the pixels under kernel area and replace the central element. This is done by the function cv2.blur().
2. Show the result using matplotlib.
3. Perform Gaussian Blurring using function cv2.GaussianBlur(). We should specify the width and height of kernel which should be positive and odd. We also should specify the standard deviation in X and Y direction, sigmaX and sigmaY respectively. If only sigmaX is specified, sigmaY is taken as same as sigmaX. 
4. Show the result using matplotlib.
5. Explain Image Smooting process.

```python
blur = cv2.blur(img,(11,11))
plt.imshow(blur)
# cv2.imshow('babyyodaUpav', blur)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
gblur = cv2.GaussianBlur(img,(11,11),1)
plt.imshow(gblur)
```

```python
titles = ['Original Image','Average','Gaussian']
images = [img[190:240,160:210],
          blur[190:240,160:210], 
          gblur[190:240,160:210]]

plt.subplots(1,3,figsize=(15,15))
for i in range(3):
    plt.subplot(1,3,i+1),plt.imshow(images[i])
    plt.title(titles[i])
plt.show()
```

<h3 style="color: gray">Image smoothing. <br>
    In Average smoothing there is a box aroud pixel and the new value of the pixel is the average of all the pixels within the box <br>
    In the Gaussian smoothing there is a Gaussian box and the new value is the average of all the pixels within the box each multiplied to the value of the corresponding box value</h3>
